```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import correlate, convolve
```

```python
X = np.random.randn(1000)
C = correlate(X, X)
plt.plot(C)
```

```python
Y = np.empty(1000)
Y[:900] = X[100:]
Y[900:] = X[:100]
C2 = correlate(X, Y)
plt.plot(C2)
```

```python
C3 = convolve(X, np.flip(Y, axis=0))
plt.plot(C3)
```

```python
(C2 == C3).all()
```
