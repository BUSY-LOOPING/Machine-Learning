```python
import tensorflow as tf
import numpy as np
import matplotlib
# matplotlib.use('nbagg')
import matplotlib.pyplot as plt

from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import Input, GlobalMaxPooling2D, Dense, Dropout, Conv2D, Flatten
from tensorflow.keras.callbacks import TensorBoard, EarlyStopping
```

```python
# label mapping
 
labels = '''airplane automobile bird cat deer dog frog horse ship truck'''.split()
```

```python
(X_train, Y_train), (X_test, Y_test) = tf.keras.datasets.cifar10.load_data()
```

### Pre-processing the data

```python
X_train.mean()
```

```python
X_train, X_test = X_train / 255 , X_test / 255
X_train.mean()
```

```python
Y_train, Y_test = Y_train.flatten(), Y_test.flatten()
```

### Building the model

```python
X_train.shape
```

```python
i = Input(shape = X_train[0].shape)  #here shape of each sample is 32 x 32 x 3
x = Conv2D(32, (3, 3), strides = 2, activation = 'relu') (i)
x = Conv2D(64, (3, 3), strides = 2, activation = 'relu') (x)
x = Conv2D(128, (3, 3), strides = 2, activation = 'relu') (x)
x = Flatten() (x)
x = Dropout(0.2) (x)
x = Dense(1024, activation = 'relu') (x)
x = Dropout(0.2) (x)
x = Dense(len(labels), activation= 'softmax') (x)

model = Model(i, x)
```

### Compiling the model

```python
model.compile(optimizer='adam', metrics=['accuracy'], loss = 'sparse_categorical_crossentropy')
```

```python
logdir = 'cifar10_logs'
tensorboard_callback = TensorBoard(log_dir = logdir)
early_stopping = EarlyStopping(patience=5,)
```

### Training the model

```python
hist = model.fit(X_train, 
                 Y_train, 
                 callbacks=[tensorboard_callback, early_stopping], 
                 validation_data=(X_test, Y_test), 
                 epochs = 40)
```

```python
plt.plot(hist.history['loss'], label = 'loss')
plt.plot(hist.history['val_loss'], label = 'val_loss')
plt.legend()
plt.show()

plt.figure()
plt.plot(hist.history['accuracy'], label = 'accuracy')
plt.plot(hist.history['val_accuracy'], label = 'val_accuracy')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import confusion_matrix
import itertools

def plot_confusion_matrix(cm, classes, normalize = False, title='Confusion matrix', cmap = plt.cm.Blues) :
    """
    This function prints and plots the confusion matrix
    """
    if normalize :
        cm = cm.astype('float') / cm.sum(axis = 1)[:, np.newaxis]
        print('Normalized Confusion matrix')
    else:
        print(cm)
        
    plt.imshow(cm, interpolation='nearest', cmap = cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)
    
    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2
    
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])) :
        plt.text(j, i, format(cm[i, j], fmt),
                horizontalalignment = 'center',
                color = 'white' if cm[i, j] > thresh else 'black')
    plt.tight_layout()
    plt.ylabel('True Label')
    plt.xlabel('Predicted label')
    plt.show()
```

```python
P_test = model.predict(X_test).argmax(axis = 1)
cm = confusion_matrix(Y_test, P_test)
plot_confusion_matrix(cm, list(range(len(labels))), normalize=False)
```

```python
misclassified_idx = np.where(P_test != Y_test)[0]
i = np.random.choice(misclassified_idx)
plt.figure()
plt.imshow(X_test[i], cmap = 'gray')
plt.title('True label: %s Predicted : %s' % (labels[Y_test[i]], labels[P_test[i]]))
plt.show()
```
