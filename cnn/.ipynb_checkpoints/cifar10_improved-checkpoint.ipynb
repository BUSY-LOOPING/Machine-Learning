```python
import tensorflow as tf
import numpy as np
import matplotlib
# matplotlib.use('nbagg')
import matplotlib.pyplot as plt

from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import Input, GlobalMaxPooling2D, Dense, Dropout, Conv2D, Flatten, BatchNormalization, MaxPooling2D
from tensorflow.keras.callbacks import TensorBoard, EarlyStopping
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
from sklearn.metrics import confusion_matrix
import itertools

def plot_confusion_matrix(cm, classes, normalize = False, title='Confusion matrix', cmap = plt.cm.Blues) :
    """
    This function prints and plots the confusion matrix
    """
    if normalize :
        cm = cm.astype('float') / cm.sum(axis = 1)[:, np.newaxis]
        print('Normalized Confusion matrix')
    else:
        print(cm)
        
    plt.imshow(cm, interpolation='nearest', cmap = cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)
    
    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2
    
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])) :
        plt.text(j, i, format(cm[i, j], fmt),
                horizontalalignment = 'center',
                color = 'white' if cm[i, j] > thresh else 'black')
    plt.tight_layout()
    plt.ylabel('True Label')
    plt.xlabel('Predicted label')
    plt.show()
```

```python
# label mapping
 
labels = '''airplane automobile bird cat deer dog frog horse ship truck'''.split()
```

```python
(X_train, Y_train), (X_test, Y_test) = tf.keras.datasets.cifar10.load_data()
```

### Pre-processing the data

```python
X_train, X_test = X_train / 255 , X_test / 255
X_test.shape
```

```python
Y_train, Y_test = Y_train.flatten(), Y_test.flatten()
Y_train.shape
```

### Building the model

```python
i = Input(shape = X_train[0].shape, name = 'input_layer')  #here shape of each sample is 32 x 32 x 3

x = Conv2D(32, (3, 3), activation = 'relu', padding = 'same', name = 'Conv_1') (i)
x = BatchNormalization(name = 'BN1') (x)
x = Conv2D(32, (3, 3), activation = 'relu', padding = 'same', name = 'Conv_2') (x)
x = BatchNormalization(name = 'BN2') (x)
x = MaxPooling2D((2, 2), name = 'MaxPooling_1') (x)

x = Conv2D(64, (3, 3), activation = 'relu', padding = 'same', name = 'Conv_3') (x)
x = BatchNormalization(name = 'BN3') (x)
x = Conv2D(64, (3, 3), activation = 'relu', padding = 'same', name = 'Conv_4') (x)
x = BatchNormalization(name = 'BN4') (x)
x = MaxPooling2D((2, 2), name = 'MaxPooling_2') (x)

x = Conv2D(128, (3, 3), activation = 'relu', padding = 'same', name = 'Conv_5') (x)
x = BatchNormalization(name = 'BN5') (x)
x = Conv2D(128, (3, 3), activation = 'relu', padding = 'same', name = 'Conv_6') (x)
x = BatchNormalization(name = 'BN6') (x)
x = MaxPooling2D((2, 2), name = 'MaxPooling_3') (x)

x = Flatten(name = 'flatten') (x)
x = Dropout(0.2, name = 'dropout1') (x)
x = Dense(1024, activation = 'relu', name = 'dense_1') (x)
x = Dropout(0.2, name = 'dropout2') (x)
x = Dense(len(labels), activation= 'softmax', name = 'output_layer') (x)

model = Model(i, x)
```

```python
model.summary()
```

### Compiling the model

```python
model.compile(optimizer='adam', metrics=['accuracy'], loss = 'sparse_categorical_crossentropy')
```

```python
logdir = 'cifar10_improved_logs'
tensorboard_callback = TensorBoard(log_dir = logdir)
early_stopping = EarlyStopping(patience=5,)
```

```python
batch_size = 32
steps_per_epoch = X_train.shape[0] // batch_size
```

```python
data_generator = ImageDataGenerator(
    width_shift_range=0.1,
    height_shift_range=0.1,
    horizontal_flip=True,
    zoom_range=0.1,
#     brightness_range=[(0, 0,5), (0.5, 1)],
)

train_generator = data_generator.flow(X_train, Y_train, batch_size)
test_generator = data_generator.flow(X_test, Y_test)
```

### Training the model

```python
hist1 = model.fit(
    X_train,
    Y_train,
#     steps_per_epoch=steps_per_epoch,
    epochs=50,
    validation_data=(X_test, Y_test),
    callbacks=[early_stopping], 
    shuffle=True
)
```

```python
plt.plot(hist1.history['loss'], label = 'loss')
plt.plot(hist1.history['val_loss'], label = 'val_loss')
plt.legend()
plt.show()

plt.figure()
plt.plot(hist1.history['accuracy'], label = 'accuracy')
plt.plot(hist1.history['val_accuracy'], label = 'val_accuracy')
plt.legend()
plt.show()
```

```python
P_test = model.predict(X_test).argmax(axis = 1)
cm = confusion_matrix(P_test, Y_test)
plot_confusion_matrix(cm, list(range(len(labels))), normalize=False)
```

```python
hist2 = model.fit(
    train_generator,
    steps_per_epoch= steps_per_epoch,
    epochs = 50,
    callbacks=[tensorboard_callback, early_stopping], 
    validation_data= (X_test, Y_test),
    shuffle = True,
)
```

```python
model.save('cifar10_improved.h5')
```
