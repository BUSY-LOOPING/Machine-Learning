```python
import numpy as np
import matplotlib
matplotlib.use('nbagg')
import matplotlib.pyplot as plt
```

```python
N = 50 #no of data points

X = np.linspace(0, 10, N)
Y = 0.5 * X + np.random.randn(N)

Y[-1] += 30
Y[-2] += 30

plt.scatter(X, Y, s= 10)
plt.show()
```

```python
X = np.vstack([np.ones(N), X]).T

#max likelihood solution
w_ml = np.linalg.solve(X.T.dot(X), X.T.dot(Y))
yhat_ml = X.dot(w_ml)
```

```python
plt.scatter(X[:,1], Y, s = 10)
plt.plot(X[:,1], yhat_ml, color='red')
plt.show()
```

```python
l2 = 1000.0
w_map = np.linalg.solve(l2 * np.eye(2) + X.T.dot(X), X.T.dot(Y))
yhat_map = X.dot(w_map)
```

```python
plt.scatter(X[:,1], Y, s = 10, label = 'data points')
plt.plot(X[:,1], yhat_ml, color='red', label = 'maximum likelihood')
plt.plot(X[:,1], yhat_map, color='green', label='map')
plt.legend()
plt.show()
```
