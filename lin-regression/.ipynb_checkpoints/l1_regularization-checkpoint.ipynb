```python
import numpy as np
import matplotlib
matplotlib.use('nbagg')
import matplotlib.pyplot as plt
```

```python
N = 50
D = 50
```

```python
X = (np.random.random((N, D)) - 0.5) * 10  # X is centred around 0 with values ranging from -5 to +5
X
```

```python
true_w = np.array([1, 0.5, -0.5] + [0] * (D - 3))  # last 3 features do not influence the output
true_w
```

```python
Y = X.dot(true_w) + np.random.randn(N) * 0.5 
```

```python
#gradient descent
costs = []

w = np.random.randn(D) / np.sqrt(D)
learning_rate = 0.001
l1 = 10

for t in range(500) : 
    yhat = X.dot(w)
    delta = yhat - Y
    w = w - learning_rate * (X.T.dot(delta) + l1*np.sign(w))
    
    mse = delta.dot(delta) / N
    costs.append(mse)
    
plt.plot(costs)
plt.show()

plt.figure()
plt.plot(true_w, label = 'true w')
plt.plot(w, label = 'w_map')
plt.legend()
plt.show()
```
