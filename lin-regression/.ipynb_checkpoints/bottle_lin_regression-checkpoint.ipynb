### Between temp and salinity

```python
temp = df[['T_degC', 'Salnty', 'Depthm']]
df_without_na = temp.dropna(how='any')

# X = [list(df_without_na[['T_degC', 'Depthm']].values)]
# X.append([np.ones(len(X[0]))])
# X = np.array(X, dtype = 'object')
# X = np.vstack(X).T

X = df_without_na[['T_degC']]
X['ones'] = np.ones(X.shape[0], dtype = 'uint8')
X = X.values


Y = df_without_na.Salnty.values
```

```python
def get_train_test(X, Y, percent = 80) :
    assert len(X) == len(Y)
    
    c = list(zip(X, Y))
    np.random.shuffle(c)

    a, b = zip(*c)

    idx = int(percent * len(X) / 100)

    X_train = np.array(a[:idx])
    X_test = np.array(a[idx:])

    Y_train = np.array(b[:idx])
    Y_test = np.array(b[idx:])
    
    return X_train, X_test, Y_train, Y_test
```

```python
def show_scatter(X, Y) :
    plt.figure()
    plt.scatter(X, Y, s = 10)
    plt.show()
    
```

```python
def fit(X, Y) :
    return np.linalg.solve(X.T.dot(X), X.T.dot(Y))

def fit_l2 (X, Y, l2) :
    return np.linalg.solve(l2 * np.eye(X.shape[1], dtype = 'uint8') + X.T.dot(X), X.T.dot(Y))

def get_yhat_1d (X, Y) :
    denominator = X.dot(X) - X.mean() * X.sum()
    a = (X.dot(Y) - Y.mean() * X.sum())/ denominator
    b = (Y.mean() * X.dot(X) - X.mean() * X.dot(Y))/ denominator
    yhat = a * X + b
    return yhat


def get_mse(Y, yhat) :
    d1 = Y - yhat 
    print(d1.dot(d1) / len(Y))
    return d1.dot(d1) / len(Y)

def make_poly(X, deg) :
    n = len(X)
    data = [np.ones(n)]
    for d in range(deg) :
        data.append(X**(d+1))
        
    return np.vstack(data).T


def show_train_vs_test_curve(X, Y, percent = 80) :

    X_train, X_test, Y_train, Y_test = get_train_test(X, Y, percent)
    
    #for training
    w_train = fit(X_train, Y_train)
    yhat_train = X_train.dot(w_train) 
#     yhat_train = get_yhat_1d(X_train[:, 0], Y_train)
    
    mse = get_mse(Y_train, yhat_train)
    
    w_train_map = fit_l2(X_train, Y_train, l2 = 1000.0)
    yhat_train_map = X_train.dot(w_train_map) 
    mse_map = get_mse(Y_train, yhat_train_map)
    
    print('Accuracy of maximum likelihood {:.2f}'.format(mse * 100))
    print('Accuracy of map {:.2f}'.format(mse_map * 100))

    plt.scatter(X_train[:,0], Y_train, c='green', s = 5, alpha=0.5)
    plt.plot(X_train[:,0], yhat_train, linewidth = 1, color = 'red', label = 'maximum likelihood')
    plt.plot(X_train[:,0], yhat_train_map, linewidth = 1, color = 'green', label = 'map')
    plt.legend()
    plt.show()
```

```python
if __name__ == '__main__' :
    show_train_vs_test_curve(X, Y, percent=50)
```
